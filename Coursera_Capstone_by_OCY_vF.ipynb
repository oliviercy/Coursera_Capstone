#### Load a custom version of the GeoJSON file, with modified centers for the 12th and 16th arrondissements

```python
geo = json.load(open("arrondissements_custom.geojson", encoding='utf-8-sig'))
```

#### Put the data in a dataframe

```python
# Aggregate properties of each arrondissement/neighborhood in a list

neighborhoods_data = []

for i in range(0,20):
    neighborhoods_data.append(i)
    neighborhoods_data[i] = geo["features"][i]["properties"]
    
# define the dataframe columns
column_names = ['Number', 'Latitude', 'Longitude']

# This code scans the properties of the arrondissements/neighborhoods
arr_lat = []
arr_lon = []
arr_list = []

for data in neighborhoods_data:
    arr_numero = data["c_arinsee"]
    arr_coordinates = data["geom_x_y"]
    arr_lat = arr_coordinates[0]
    arr_lon = arr_coordinates[1]

    arr_list.append([arr_numero, arr_lat, arr_lon])

neighborhoods = pd.DataFrame(arr_list,columns=column_names) # Create a dataframe with number + Lat/Lon
neighborhoods.sort_values(by=['Number'], inplace=True) # Sort the dataframe by number from 01 to 20
neighborhoods.reset_index(drop = True, inplace = True) # Reindex

# https://stackoverflow.com/questions/45747589/copying-a-column-from-one-dataframe-to-another-gives-nan-values
# Verify if indexes match between `neighborhoods` and `arrondissements` dataframes
#neighborhoods.index.equals(arrondissements.index)
```

```python
arrondissements[['Latitude', 'Longitude']] = neighborhoods[['Latitude', 'Longitude']]
arrondissements
```

#### Use the GeoJSON polygons to draw the <span style="color:Crimson">borders</span> of the arrondissements and place markers on the map. Let's also visualize the <span style="color:Crimson">density</span> of each arrondissement, based on dataframe `arrondissements`

```python
# https://stackoverflow.com/questions/35516318/plot-colored-polygons-with-geodataframe-in-folium

# Add the borders of each arrondissement
folium.GeoJson(
    geo,
    lambda x :{'fillColor':'yellow', 'fillOpacity':0, 'color':'#3973ac', 'weight':2, 'opacity':1}
    ).add_to(paris)

# help(folium.Icon)
# Icons gallery: https://fontawesome.com/icons?d=gallery

# Put markers on each arrondissement
for i in range(0,20):
    prop = geo["features"][i]["properties"]
    popup_string = '{} - {}'.format(prop["l_ar"], prop["l_aroff"])
    folium.Marker(
        prop["geom_x_y"],
        popup = popup_string,
        icon=folium.Icon(color='cadetblue', icon='home', prefix='fa'),
    ).add_to(paris)

# Density map based on 'arrondissements' dataframe
paris.choropleth(
 geo_data=geo,
 name='choropleth',
 data=arrondissements,
 columns=['objectid', 'Dens. 2017'],
 key_on='feature.properties.objectid',
 fill_color='YlGnBu',   #Color schemes: https://github.com/python-visualization/folium/blob/v0.2.0/folium/utilities.py#L104
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Population Density per Arrondissement of Paris',
 highlight=True
)
folium.LayerControl().add_to(paris)

paris
```

From the Foursquare lab we did during the course, we know that all the information is in the *items* key. Before we proceed, let's borrow the **get_category_type** function from the Foursquare lab.

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

Now we are ready to clean the json and structure it into a *pandas* dataframe.

```python
#venues = results['response']['groups'][0]['items']
venues = results['response']['venues']

nearby_venues = pd.json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['name', 'categories', 'location.postalCode', 'location.lat', 'location.lng']
nearby_venues = nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
```

### B. Explore Neighborhoods in Paris

#### Let's create a function to repeat the same process to all the neighborhoods in Manhattan

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=100):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create URL
        url2 = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&categoryId={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION,
            categoryId,
            lat,
            lng,
            radius,
            LIMIT)
            
        # make the GET request
        results2 = requests.get(url2).json()['response']['venues']

        # return only relevant information for each nearby venue
        for i in range(0,len(results2)):
            venues_list.append([(
                name, 
                lat,
                lng, 
                results2[i]['name'], 
                results2[i]['location']['lat'], 
                results2[i]['location']['lng'],
                results2[i]['categories'][0]['name'])])

        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Neighborhood',
                  'Neighborhood Latitude',
                  'Neighborhood Longitude',
                  'Venue',
                  'Venue Latitude',
                  'Venue Longitude',
                  'Venue Category']
          
    return(nearby_venues)
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

#### Problem: the `surface` shares its CSS class with others criteria and is not always at the same place on the .HTM page

The block below may take up to 30 seconds to execute (loops through lots of data)

```python
# Retrieve the surface indicated in each offer.
# One example of criteria block here: https://pasteboard.co/J7KfY3G.png
# Another example of criteria block: https://pasteboard.co/J7KgwBq.png

surfacez = []
surfaces = []
sub = 'm²'

# Loop in each offer
for i in range (0,140):
    with open(statics[i], "r") as filez:
        filez = filez.read()
        sursoup = BeautifulSoup(str(filez),'html.parser')

        # Look for the criteria block
        lasurface = sursoup.find_all(class_='_3eNLO _137P- P4PEa _35DXM')
        tagz = []

        # Put the criteria in `lasurface` list: house/apartment, surface, number of bedrooms, greenhouse gazes index...
        for tag in lasurface:
            # https://stackoverflow.com/questions/13779526/finding-a-substring-within-a-list-in-python
            tagz.append(tag.text.strip())

        # https://stackoverflow.com/questions/13779526/finding-a-substring-within-a-list-in-python
        # Detect the 'm²' string in the list of criteria to locate the surface
        surfacez.append([s for s in tagz if sub in s])
        if not surfacez[i]:        # Check if the surface exists in the `surfacez` list, hence in this offer
            surfaces.append('NaN')
        elif not surfacez[i][0]:   # Check if the surface exists in the `surfacez` list, hence in this offer
            surfaces.append('NaN')
        else:
            surfaces.append(surfacez[i][0]) # If the surface exists in this offer, append the `surfaces` list
```

```python
# https://stackoverflow.com/questions/57132343/pandas-numpy-groupby-aggregate-inc-integer-mean-filter
mean_price_arr = df_offers.groupby('Num.')['Price_per_m²'].aggregate(['min', max, np.mean, np.median, 'count']).sort_values('mean', ascending=False).astype(np.int32)
mean_price_arr
```
